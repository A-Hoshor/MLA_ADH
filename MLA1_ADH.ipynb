####Answers 9-11

```python
def convert_temp(temp_c):
  fahrenheit=float
  fahrenheit= (temp_c * 1.8) + 32
  return fahrenheit
```

```python
df['temp_f'] = df['temp_c'].apply(convert_temp)
```

```python
df.sample(5)
```

```python
df = df[['date_time','temp_c','temp_f','dew_point','relative_humidity','wind_speed','visibility','pressure','weather']]
```

####Answer 13

```python
month_df = df[['date_time','temp_f']].copy() #create new data frame with only the date and temperature
```

```python
month_df = month_df.groupby([month_df['date_time'].dt.year, df['date_time'].dt.month]).mean()
```

```python
month_avg = month_df['temp_f'] #convert from data frame to data series
```

```python
month_avg
```
